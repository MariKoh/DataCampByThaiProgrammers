```
#NP Array Vs List

#List
cent = [175, 185, 168, 177]
inch = []

for i in cent:
  inch.append(round(i/2.54, 2))

print(inch)

#List Comprehension
inch2 = [round(i/2.54, 2) for i in cent]

print(inch2)


#Map Lambda
inch3 = list(map(lambda x: round(x/2.54, 2), cent))
print(inch3)



#Numpy Array
#Vectorization
inch4 = (np.array(cent) / 2.54).round(2)
print(inch4)
```
