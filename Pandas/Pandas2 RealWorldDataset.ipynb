#Data Cleaning

##Missing Values
ต้องทำให้เป็น NaN ก่อน

```
Data.dropna() #Default axis=0 ลบ Rows
# Data.dropna(axis=1) วิธีนี้ไม่แนะนำ เพราะหายไปทั้ง Columns
```

##แทนค่า NaN ด้วย 0

```
Data.fillna(value=(int(0))) #แบบนี้ไม่ค่อยดีเพราะบาง Column ที่เป็น String การใส่ 0 มันจะไม่ Make Sense
```

```
Data['normalized-losses'].fillna(value=0)
```

```
avg = Data['normalized-losses'].mean()
avg
```
