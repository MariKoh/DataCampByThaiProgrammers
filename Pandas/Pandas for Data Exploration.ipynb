#Crosstab

```
pd.crosstab(index=Data['drv'],
            columns=Data['class'],
            margins=True, #สร้าง Columns sum ขึ้นมาข้างหลัง
            margins_name='Total',
            normalize=True) #ทำให้มันเป็นอัตราส่วน %
```

```
pd.crosstab(index=Data['drv'],
            columns=Data['class'],
            margins=True, #สร้าง Columns sum ขึ้นมาข้างหลัง
            margins_name='Total',
            normalize=True)*100 #ทำให้มันเป็นอัตราส่วน %
```

```
(pd.crosstab(index=Data['drv'],
            columns=Data['class'],
            margins=True, #สร้าง Columns sum ขึ้นมาข้างหลัง
            margins_name='Total',
            normalize=True)*100).round(1) #ทำให้มันเป็นอัตราส่วน %
```

```
Data['Trans_AM'] = Data['trans'].str[0]
Data['Trans_AM'] = Data['Trans_AM'].str.upper()
Data
```

```
pd.crosstab(index=Data['Trans_AM'],
            columns=Data['manufacturer'],
            # rownames='Transmission',
            # colnames='Brand',
            margins=True,
            margins_name='Total'
            )
```

```
#Multilavel
pd.crosstab(index=[Data['class'], Data['Trans_AM']],
            columns=[Data['fl'], Data['manufacturer']],
            # rownames='Transmission',
            # colnames='Brand',
            margins=True,
            margins_name='Total')
```

```
#Multilavel
#ใช้ Mean แทน Sum
pd.crosstab(index=[Data['class'], Data['Trans_AM']],
            columns=[Data['fl'], Data['manufacturer']],
            values=Data['cty'],
            aggfunc=np.mean,
            # rownames='Transmission',
            # colnames='Brand',
            margins=True,
            margins_name='Total')
```

```
#ได้ค่าออกมาเป็น DataFrame เราก็สามารถใช้ Function ของ DataFrame ได้
pd.crosstab(index=[Data['class'], Data['Trans_AM']],
            columns=[Data['fl'], Data['manufacturer']],
            values=Data['cty'],
            aggfunc=np.mean,
            # rownames='Transmission',
            # colnames='Brand',
            margins=True,
            margins_name='Total').sort_index().loc[['compact', 'minivan']]
```

```
#ได้ค่าออกมาเป็น DataFrame เราก็สามารถใช้ Function ของ DataFrame ได้
pd.crosstab(index=[Data['class'], Data['Trans_AM']],
            columns=[Data['fl'], Data['manufacturer']],
            values=Data['cty'],
            aggfunc=np.max,
            # rownames='Transmission',
            # colnames='Brand',
            margins=True,
            margins_name='Total').sort_index().loc[['compact', 'minivan']]
```

#คำนวนข้อมูลหุ้น โดย Groupby

```
from pandas_datareader import data
```

##Get Stock

```
stocks = pd.Series(['scc', 'ptt', 'aot']) + '.bk'
stocks
```

```
df  = data.DataReader(name=stocks,
                      data_source='yahoo',
                      start='2020-01-02',
                      end='2020-1-11')
df
```

##Method 1: Merge Data Frame

```
ds = df.unstack()['Close'].reset_index().rename(columns={0: 'Close'})
ds
```

```
d_first = ds.groupby(by='Symbols').first() #เอามาเฉพาะตัวแรก
d_first
```

```
dx = ds.merge(right=d_first,
              on='Symbols',
              suffixes=('', '_1'))
dx
```

```
dx['chg'] = dx['Close'] - dx['Close_1']
dx['Pct_Chg'] = ((dx['Close'] / dx['Close_1'] - 1) * 100).round(2)
dx
```

```
dx.drop(columns='Date_1',
        inplace=True)
dx
```

```
dx.rename(columns={'chg': 'Chg'},
          inplace=True)
dx
```

```
sns.relplot(x='Date',
            y='Close',
            hue='Symbols',
            kind='line',
            data=dx,
            marker='o');
```

```
sns.relplot(x='Date',
            y='Pct_Chg',
            hue='Symbols',
            kind='line',
            data=dx,
            marker='o');
```

```
sns.relplot(x='Date',
            y='Pct_Chg',
            hue='Symbols',
            kind='line',
            data=dx,
            marker='o',
            col='Symbols',
            col_wrap=3);
```
