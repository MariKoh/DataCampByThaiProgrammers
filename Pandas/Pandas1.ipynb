```
#Multi Index
level1 = ['รถญี่ปุ่น', 'รถญี่ปุ่น', 'รถญี่ปุ่น', 'รถยุโรป', 'รถยุโรป', 'รถยุโรป']
level2 = ['ฮอนด้า', 'โตโยต้า', 'นิสสัน', 'เบ๊นซ์', 'บีเอ็ม', 'พอร์เช่']
cols = ['ยอดขายปี 2018 (Mb)', 'ยอดขายปี 2019 (Mb)']

print(level1)
print(level2)
print(cols)

index = list(zip(level1, level2))
print(index)

revenue = np.random.randint(5, 50, (6, 2))
print(revenue)
```

#Series

```
Score = [90, 80, 69, 79, 40]
Name = ['MariKoh', 'John ชาวไร่', 'John Wick', 'John Snow', 'Johnny']

S = pd.Series(data=Score, index=Name, dtype=int, name='Students Name')
print(S)
```

```
Score2 = np.array(Score)
Name2 = np.array(Name)

S2 = pd.Series(index = Name2,
               data = Score2,
               dtype = int,
               name = 'Students Name')

print(type(Score2))
print(type(Name2))
print(S2)
```

```
print(f'Series Value = {S2.values}')
print(f'Series Index = {S2.index}')
```

```
print(S2[:2])
```

```
print(S2[[0, 2, 4]]) #Slice ค่าเดียว
```

```
print(S2[::2])
```
