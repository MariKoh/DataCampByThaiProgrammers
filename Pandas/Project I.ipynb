#Basic Programing

1

```
num = float(input('Please enter a number: '))
lst = []

while num != 0:
  lst.append(num)
  num = float(input('Please enter a number: '))

if num == 0:
  print(f'ค่าเฉลี่ยของตัวเลขที่กรอกเข้ามา = {np.mean(lst)}')
```

2

```
Low = int(input('Lowest Range:'))
High = int(input('Highest Range: '))


for i in range(Low, High+1):
  if i > 1:
    for j in range(2, i):
      if i % j == 0:
        break
    else:
      print('%d is a prime number.'%i)

```
