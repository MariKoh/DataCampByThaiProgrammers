####3.3.4.1 Set Union

```
setA = {1, 2, 3, 4, 5}
setB = {3, 4, 5, 6, 7}

Union = setA | setB
print('a union b =',Union)

intersection = setA & setB
print('a intersect b =', intersection)

DifferenceAB = setA - setB
DifferenceBA = setB - setA
print('Difference setA and setB =', DifferenceAB)
print('Difference setB and setA =', DifferenceBA)

Symmetry_Difference =  Union - intersection
Symmetry_Difference2 = (setA - setB) | (setB - setA) #a-b union b-a
Symmetry_Difference3 = setA ^ setB
print('Symmetry Difference =', Symmetry_Difference)
print('Symmetry Difference2 =', Symmetry_Difference2)
print('Symmetry Difference3 =', Symmetry_Difference3)
```
