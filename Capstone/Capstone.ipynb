#Part1 (Basic Programming)

```
num = float(input('Plz enter a number: '))
lst = []
s = 0

while num != 0:
  lst.append(num)
  num = float(input('Plz enter a number: '))
  print('Press "0" in case you want to exit.')

for i in lst:
  s = s + i

print(
f'\n\
The list contains {lst}, there are {len(lst)} numbers inside.\n\
Sum of the list is {s} and the average of this list is {round(s/len(lst), 2)}.\n\
Finished!!'
)
```

```
Lowest = int(input('Plz enter the lowest number: '))
Highest = int(input('Plz enter the highest number: '))
lst = []

for i in range(Lowest, Highest+1):
  if i > 1:
    for j in range(2, i):
      if i % j == 0: #หาว่าอะไรไม้ใช่ Prime Number
        # lst.append(j)
        break
    else: #Else คือ Prime Number
      lst.append(i)
print(f'Numbers from {Lowest} - {Highest} have {lst} as a prime number, which has total of {len(lst)} numbers.')
```

```
low_bar = int(input('Plz enter the lower bound: '))
high_bar = int(input('Plz enter the higher bound: '))

for number in range(low_bar, high_bar+1):
  if number > 1:
    for i in range(2, number):
      if number % i == 0:
        break
    else:
      print(f'{number} is a prime number.')
```

```
not_prime = [j for i in range(2,10) for j in range(i*2,100,i)]

prime = [x for x in range(2,100) if x not in not_prime]

print(prime)
```

```
check = int(input('Plz enter a number: '))

if check > 1 and check != 2 and check != 3 and check != 5 and check != 7:
  if check % 2 == 0 or check % 3 == 0 or check % 5 == 0 or check % 7 == 0:
    print(f'{check} is not a prime number.')
  else:
    print(f'{check} is a prime number.')
else:
  print('1, 2, 3, 5, 7 is not a prime number.')
```
