#Import Libraries

```
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import files
import io
```

#Plotly
https://plotly.com/python/pie-charts/

```
rows = ['MariKoh', 'BieBird', 'Rathy', 'Beam']
cols = ['Quiz1', 'Mid-Term', 'Quiz2', 'Final']
vals = np.random.randint(low=40,
                         high=100,
                         size=(4, 4))

Data = pd.DataFrame(data=vals,
                    index=rows,
                    columns=cols)

Data
```

```
fig = px.bar(data_frame=Data, #ใช้ Data จากไหน
             x=Data.index, #แกน X คืออะไร
             y='Final', #แกน y คืออะไร
             hover_data=['Quiz1', 'Mid-Term','Quiz2'], #เวลาเอา Mouse ชี้ที่กราฟ จะให้มีข้อมูลอะไรเพิ่มมาบ้าง
             color='Final', #ให้เรียงสีจาก Columns ไหน #ควรใส่ให้ตรงกับข้อมูลที่ Plot ในแกน y เพื่อลดความหลอน
             labels={'x': 'ชื่อนักเรียน', #เปลี่ยนชื่อแต่ละแกนเป็นอะไร
                     'Final': 'Final Score'},
             height=500, #กำหนดความสูงของ Figure
             title='Students Final Score', #ใส่ชื่อของกราฟนี้
             width=500) #กำหนดความกว้างของ Figure

fig.update_layout(xaxis_tickangle=-45) #เปลี่ยนชื่อนักเรียนเป็น ตัวเอียง #เวลาสร้างกราฟ ใส่ตัวแปร fig จะทำให้ทำอะไรได้ง่ายกว่า
fig.show()
```

#Read Data

```
Uploaded = files.upload()
```

```
Data = pd.read_csv(io.StringIO(Uploaded['Automobile_data.csv'].decode('utf-8')),
                   na_values={
                       'normalized-losses': ['?', '0', 0],
                       'stroke': ['?', '0', 0],
                       'horsepower': ['?', '0', 0],
                       'peak-rpm': ['?', '0', 0],
                       'price': ['?', '0', 0],
                       'average-mileage': ['?', '0', 0]
                   })

Data['price'] = pd.to_numeric(Data['price'])
Data['horsepower'] = pd.to_numeric(Data['horsepower'])

Data.head()
```

```
Data2 = Data.groupby('horsepower').mean()
Data2.tail()
```

#Line Graph

```
fig = px.line(data_frame=Data,
              x='horsepower',
              y='price',
              title='Average Price for Each Horsepower Range.')

fig.show()
```

```
fig2 = px.line(data_frame=Data2,
               x=Data2.index,
               y='price',
               title=('Horse Power'),
               labels={'x': 'HorsePower Range',
                       'price': 'Price Range'},
               height=500,
               width=500)

fig2.show()
```

```
df_ex = px.data.gapminder().query('continent=="Oceania"')
df_ex
```

```
fig3 = px.line(data_frame=df_ex[df_ex['country'] == 'Australia'],
               x='year',
               y='pop',
               hover_data=df_ex[['gdpPercap',
                                 'lifeExp']],
               title='Australian Population Growth',
               height=450,
               width=800,
               labels={'year': 'Year',
                      'pop': 'Australian Population'})
fig3.show()
```

```
fig4 = px.line(data_frame=df_ex[df_ex['country'] == 'New Zealand'],
               x='year',
               y='pop',
               hover_data=df_ex[['gdpPercap']], #เวลาทำอะไรที่มันสามารถใส่ได้แค่ค่าเดียว ให้ใส่ [] ไว้ด้วย
               title='New Zealand Population Growth',
               labels={'year': 'Year',
                       'pop': 'Population'},
               height=450,
               width=800,
               line_shape='linear')

fig4.show()
```

```
fig5 = px.line(data_frame=df_ex,
               x='year',
               y='lifeExp',
               color='country',
               height=450,
               width=800,
               title='Population Growth',
               labels={'year': 'Year',
                       'pop': 'Population'},
               hover_data=['gdpPercap', 
                           'pop'])

fig5.show()
```

#PieChart
เลือกสี https://plotly.com/python/builtin-colorscales/#discrete-color-sequences

```
fig = px.pie(data_frame=Data,
             values='price',
             names='make',
             title='Price Range',
             hover_data=['horsepower'],
             width=500,
             height=500)

fig.show()
```

```
Datamean = Data.groupby('make').mean()
Datamean
```

```
fig2 = px.pie(data_frame=Datamean,
              names=Datamean.index,
              values='price',
              title='Average Price for each brand',
              width=500,
              height=500,
              hover_data=Datamean[['highway-mpg', 'city-mpg']].round(0))

fig2.show()
```

```
fig3 = px.pie(data_frame=Datamean,
              names=Datamean.index,
              values='horsepower',
              height=500,
              width=500,
              title='Average HorsePower',
              color_discrete_sequence=px.colors.sequential.Blugrn) #เลือกสี https://plotly.com/python/builtin-colorscales/#discrete-color-sequences

fig3.show()
```

#Bubble Chart

```
Datahorse = Data.groupby('horsepower').mean()
Datahorse.tail()
```

```
figb = px.scatter(data_frame=Datahorse,
                  x=Datahorse.index,
                  y='price',
                  height=500,
                  width=500,
                  color='price', #การไล่สี
                  color_discrete_sequence=px.colors.cyclical.mrybm, #ใช้เปลี่ยนสี (สีเดียว)
                  size='engine-size',
                  hover_data=['city-mpg', 'highway-mpg'],
                  title='Bubble Price Chart',
                  labels={'x': 'Horse Power',
                          'price': 'Average Price'},
                  )

figb.show()
```

#Box Plot

```
figbox = px.box(data_frame=Data,
                x='make',
                y='price',
                color='fuel-type',
                labels={'make': 'Brand',
                        'price': 'Price Range'},
                title='Price Range by Brand',
                height=450,
                width=800,
                points='all') #ใส่ Stripplot มาข้างๆ Boxplot เพื่อดูการกระจายตัว

figbox.update_layout(xaxis_tickangle=45)
figbox.show()
```

#Choropleth Map

```
figmap = px.choropleth(locations=['CA', 'TX', 'NY', 'NJ', 'AL', 'CO', 'NV'],
                       locationmode='USA-states',
                       color=[1, 2, 3, 4, 5, 6, 7],
                       scope='usa')

figmap.show()
```

```
US_Data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')
```

```
US_Data.tail()
```

```
for col in US_Data.columns:
  US_Data[col] = US_Data[col].astype(str)

US_Data['Text'] = US_Data['state'] + '<br>' + \
                  'Beef = ' + US_Data['beef'] + ' Dairy = ' + US_Data['dairy'] + '<br>' + \
                  'Fruits = ' + US_Data['total fruits'] + ' Veggies = ' + US_Data['total veggies'] + '<br>' + \
                  'Corn = ' + US_Data['corn'] + ' Wheat = ' + US_Data['wheat'] + '<br>' + \
                  'Pork = ' + US_Data['pork'] + ' Total Exports = ' + US_Data['total exports'] 


figmap2 = go.Figure(data=go.Choropleth(locations=US_Data['code'],
                                       z = US_Data['pork'],
                                       locationmode='USA-states',
                                       colorscale='Picnic',
                                       colorbar_title='Millions USD',
                                       text = US_Data['Text']))

figmap2.update_layout(title_text='2011 US Exports',
                      geo_scope = 'usa')
figmap2.show()
```
