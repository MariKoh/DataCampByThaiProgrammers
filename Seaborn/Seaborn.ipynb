#Countplot

```
#อยากทราบว่า 1 Brand มีกี่คัน

sns.countplot(x=Data['make']) #ใส่ agrument แค่แกน x ตัวเดียว
```

#Boxplot

```
#แปลงจาก str เป็น numeric
#และต้อง Clean Data ให้เรียบร้อยก่อน
Data['price'] = pd.to_numeric(Data['price'])

Data
```

```
#ต้องใส่ numerical และ Categorical ทั้ง 2 อย่าง
fig = plt.figure(figsize=(16, 9))

sns.boxplot(x=Data['make'],
            y=Data['price'])

plt.title('Price Range',
          fontsize=48,
          c='#a9288c')
fig.autofmt_xdate()
```

```
fig = plt.figure(figsize=(8, 4.5))

sns.boxplot(x='body-style',
            y='price',
            data=Data)

plt.title('Different Price Different Body Style',
          c='#a9288c',
          fontsize=20)
fig.autofmt_xdate()
```

#Stripplot
คล้ายๆ Scatter Plot แต่มี Feature X มากกว่า 1 ตัว

```
fig = plt.figure(figsize=(8, 4.5))

sns.stripplot(x='body-style',
              y='horsepower',
              data=Data)

plt.grid()
fig.autofmt_xdate
```

```
plt.figure(figsize=(8, 4.5))

sns.stripplot(x='body-style',
              y='price',
              data=Data)
```

```
plt.figure(figsize=(8, 4.5))

sns.stripplot(x='drive-wheels',
              y='price',
              data=Data,
              hue='num-of-doors') #แบ่งข้อมูลภายใน Data อีกทีนึง แต่มันยังซ้อนกันอยู่ ยังไม่ค่อยรู้เรื่อง
```

```
plt.figure(figsize=(8, 4.5))

sns.stripplot(x='body-style',
              y='horsepower',
              data=Data,
              hue='num-of-doors',
              dodge=True) #แบ่งข้อมูลภายในให้แยกกันอีกครั้ง
```

#Swarmplot
คล้ายๆ Stripplot แต่จะบานออกข้างๆด้วย

```
plt.figure(figsize=(8, 4.5))

sns.swarmplot(x='body-style',
              y='horsepower',
              data=Data,
              hue='num-of-doors',
              dodge=True)
```

#HeatMap
ต้องแปลง DataFrame ให้เป็น Matrix Form

```
#Crosstab
Data2 = pd.crosstab(index=Data['body-style'],
                    columns=Data['drive-wheels'])

Data2
```

```
sns.heatmap(Data2,
            cmap='coolwarm',
            annot=True) # ใส่สีได้
```

```
#Pivot Table
Data3 = pd.pivot_table(data=Data,
                       index='body-style',
                       columns='drive-wheels',
                       values='price')

Data3
```

```
plt.figure(figsize=(8, 4.5))
sns.heatmap(data=Data3,
            cmap='coolwarm',
            annot=True,
            fmt='1.2f',
            linecolor='White', #เปลี่ยนสีเส้นเป็นสีขาว
            linewidth=5) #เส้นแบ่ง cell
```

```
#Groupby
Data4 = Data[['body-style', 'price']].dropna()

g = Data4.groupby('body-style')
bp = g.describe()['price'][['mean', 'std', 'max']]
```

```
plt.figure(figsize=(8, 4.5))

sns.heatmap(data=bp,
            cmap='coolwarm',
            annot=True,
            fmt='1.2f',
            linewidth=5,
            linecolor='#c0fff4')
```

```
#Correlation จะใช้ได้แค่ Data ที่เป็น Numerical เพราะ String มันเอามาหาค่าไม่ได้
plt.figure(figsize=(16, 9))

sns.heatmap(data=Data.corr(),
            cmap='coolwarm',
            linewidth=5,
            linecolor='#c0fff4',
            annot=True,
            fmt='1.2f')

plt.title('Data Correlation')
plt.show()
```

```
plt.figure(figsize=(20, 20))

sns.clustermap(data=Data.corr(),
               cmap='coolwarm',
               annot=True)

plt.title('Data Correlation')
plt.show()
```

#DistributionPlot
คล้ายๆ Histrogram

```
plt.figure(figsize=(8, 4.5))

sns.distplot(Data['make'].value_counts(),
             color='#f08080',
             bins=Data['make'].nunique(),
             label='Where is label')

plt.title('Brand')
plt.grid()
plt.legend()
plt.show()
```

```
plt.figure(figsize=(8, 4.5))

sns.distplot(Data['horsepower'],
             color='#f08080',
             kde=False)

plt.title('Horse Power')
plt.show()
```
