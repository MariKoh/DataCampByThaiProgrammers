#Multi Graph

```
x = np.arange(1, 21)
y = np.sin(x)

plt.subplot(2, 2, 1)
plt.title('Multiple Graph in figure')
plt.plot(x, y)

plt.subplot(2, 2, 3) #plt.subplot(อัตราส่วนแกน y, อัตราส่วนแกน x, กราฟที่เท่าไหร่)
plt.plot(y, x)
plt.show()
```

```
x = np.arange(1, 21)
y = np.sin(x)

fig = plt.figure()
ax1 = fig.add_subplot(231,
                      title='Num1')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.plot(x, y,
         label='Sin x')

ax2 = fig.add_subplot(232,
                      title='Num2')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.plot(x, x*2,
         label='2X',
         c='#B5C2FF')

ax3 = fig.add_subplot(233,
                      title='Num3')
ax3.set_xlabel('X Axis')
ax3.set_ylabel('Y Axis')
ax3.plot(x, x**2,
         label='x^2',
         c='#FF8247')

ax4 = fig.add_subplot(234,
                      title='Fig4')
ax4.set_xlabel('X Axis')
ax4.set_ylabel('Y Axis')
ax4.plot(x, x*3,
         label='3X',
         c='#009966')

ax6 = fig.add_subplot(236,
                      title='Fig6')
ax6.set_xlabel('X Axis')
ax6.set_ylabel('Y Axis')
ax6.plot(x, x**3,
         label='X^3',
         c='#fba34b')

plt.tight_layout() #ใส่ให้กราฟไม่ซ้อนกัน
plt.legend()
plt.show()
```

```
x = np.arange(1, 31)
y = np.cos(x)

fig = plt.figure(figsize=(9, 4.5))

ax1 = fig.add_axes([0.1, 0.1, 0.95, 0.85])
ax1.plot(x, x**2,
         c='#fba34b',
         label='x**2')

ax2 = fig.add_axes([0.2, 0.5, 0.3, 0.4])
ax2.plot(x, (x**3)-x,
         label='(x**3)-x',
         c='#009966')

ax1.legend(loc=5)
ax2.legend()
plt.show()
```

```
#Quiz5
x = np.arange(1, 11)
y = x**2
z = x**3

plt.figure(figsize=(9, 4.5),
           dpi=300,
           facecolor='#a6e377')

plt.plot(x, y,
         c='r',
         marker='o',
         label='X^2')
plt.plot(x, z,
         c='b',
         marker='^',
         label='X^3',
         ls='-.')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X^2 & X^3')
plt.legend()
plt.grid()
plt.show()
```

```
#Quiz6
f1x = np.arange(1, 11)
f1y = f1x**2

f2y = np.arange(1, 11)
f2x = f2y**2

plt.figure(figsize=(9, 4.5),
           dpi=100,
           facecolor='#a6e377')

plt.plot(f1x, f1y,
         c='r',
         marker='o',
         label='X^2')
plt.plot(f2x, f2y,
         c='b',
         marker='^',
         label='Y^2')

plt.title('What is this?')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid()
plt.show()
```

```
#Quiz7
y= np.arange(1, 11)
x = y**2+4*y

plt.figure(figsize=(9, 4.5),
           dpi=100,
           facecolor='#a6e377')

plt.plot(x, y,
         c='#eee869', 
         marker='p',
         label='y^2 + 4*y')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is title')
plt.grid()
plt.legend()
plt.show()
```

```
#Quiz8

f1y = np.arange(1, 11)
f1x = f1y**3

f2x = np.arange(1, 6)
f2y = f2x**2+2

fig = plt.figure(figsize=(9, 4.5),
           dpi=100,
           facecolor='#9ee1fe')

ax1 = fig.add_axes([0, 0, 0.95, 0.85]) #สร้าง Sub Figure
ax1.plot(f1x, f1y,
         c='#cf6744',
         marker='o',
         ls='-.',
         linewidth=3,
         label='Y**3')

ax2 = fig.add_axes([0.5, 0.1, 0.4, 0.4]) #สร้าง Sub Figure
ax2.plot(f2x, f2y,
         c='#463558',
         marker='v',
         ls='-',
         linewidth=3,
         label='X**2+2')

ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')

ax1.set_title('This is figure 1.')
ax2.set_title('This is figure 2.')

ax1.legend()
ax2.legend()

ax1.grid()
ax2.grid()

# plt.tight_layout()
plt.show()
```

```
#Quiz9
plt.figure(figsize=(7, 7))

p1 = Data[Data['Pclass'] == 1].shape[0]
p2 = Data[Data['Pclass'] == 2].shape[0]
p3 = Data[Data['Pclass'] == 3].shape[0]

val = [p1, p2, p3]
label = ['Class 1', 'Class 2', 'Class 3']

plt.pie(val,
        labels=label,
        shadow=True,
        explode=[0.1, .1, 0.1],
        autopct='%1.2f%%')
plt.show()
```

```
#Quiz10
S = Data[Data['Embarked'] == 'S'].shape[0]
Q = Data[Data['Embarked'] == 'Q'].shape[0]
C = Data[Data['Embarked'] == 'C'].shape[0]

em = ['Southampton', 'Queenstown', 'Cherbourg']

plt.figure(figsize=(7, 7))

plt.pie([S, Q, C],
        labels=em,
        explode=[0, 0.1, 0],
        autopct='%1.2f%%',
        shadow=True)

plt.title('Embarked')
plt.show()
```
